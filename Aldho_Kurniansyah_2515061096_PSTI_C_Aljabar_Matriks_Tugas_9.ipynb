

Subbab Scalar



```python
#buat variabel skalar x dan simpan suatu nilai skalar
x_skalar = 11
print(x_skalar)

#tampilkan tipe data dari x
type(x_skalar)

# buat variabel skalar y dan simpan suatu nilai skalar
y_skalar = 22
print(y_skalar)

#tampilkan tipe data dari y
type(y_skalar)

#jumlahkan x+y
hasil = x_skalar + y_skalar
print(f"x + y = {x_skalar} + {y_skalar} = {hasil}")

#buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = 1.1

#jumlahkan x_float dengan y
print("x_float + y = ", x_float + y_skalar)

#tampilkan tipe data dari x_float
type(x_float)
```

```python
import tensorflow as tf

# Buat variabel skalar x dan simpan suatu nilai skalar menggunakan TensorFlow
x_tf = tf.Variable(11)
print(f"Nilai x = {x_tf}")

# Tampilkan tipe data dari x_tf_skalar
print(f"Tipe data TensorFlow x_skalar= {x_tf.dtype}")

# Buat variabel skalar y dan simpan suatu nilai skalar menggunakan TensorFlow
y_tf = tf.Variable(22)
print(f"Nilai y = {y_tf}")

# Tampilkan tipe data dari y_tf_skalar
print(f"Tipe data TensorFlow y_skala = {y_tf.dtype}")

# Jumlahkan x_tf_skalar + y_tf_skalar menggunakan TensorFlow
sum_tf = tf.add(x_tf, y_tf)
print(f"TensorFlow x + y = {sum_tf}")

# Buat variabel bernama x_tf_float dan simpan suatu nilai desimal menggunakan TensorFlow
x_tf_float = tf.Variable(2.2)
print(f"TensorFlow x_float = {x_tf_float}")

#jumlahkan x_tf_float dengan y
sum_tf_float = tf.add(x_tf_float, tf.cast(y_tf, tf.float32))
print(f"TensorFlow x_float + y = {sum_tf_float}")

# Tampilkan tipe data dari x_tf_float
print(f"Tipe data TensorFlow x_float= {x_tf_float.dtype}")
```

```python
import torch

# Buat variabel skalar x dan simpan suatu nilai skalar menggunakan PyTorch
x_pt = torch.tensor(11)
print(f"Nilai x = {x_pt}")

# Tampilkan tipe data dari x_torch_skalar
print(f"Tipe data PyTorch x_skalar= {x_pt.dtype}")

# Buat variabel skalar y dan simpan suatu nilai skalar menggunakan PyTorch
y_pt = torch.tensor(22)
print(f"PyTorch y_skalar= {y_pt}")

# Tampilkan tipe data dari y_torch_skalar
print(f"Tipe data PyTorch y_skalar= {y_pt.dtype}")

# Jumlahkan x_torch_skalar + y_torch_skalar menggunakan PyTorch
sum_torch = x_pt + y_pt
print(f"PyTorch x + y = {sum_torch.item()}")

# Buat variabel bernama x_torch_float dan simpan suatu nilai desimal menggunakan PyTorch
x_torch_float = torch.tensor(2.2)
print(f"PyTorch x_float= {x_torch_float}")

# Jumlahkan x_torch_float dengan y_torch_skalar menggunakan PyTorch
print(f"PyTorch x_float + y = {x_torch_float.item()} + {y_pt.item()} = {(x_torch_float + y_pt).item()}")

# Tampilkan tipe data dari x_torch_float
print(f"Tipe data PyTorch x_float= {x_torch_float.dtype}")
```

Subbab Vectors and Vector Transposition

```python
import numpy as np

#buat vector x dengan ukuran 3x1
x_vector = np.array([[11],[22],[33]])
print("x= ", x_vector)

#hitung panjang vector x
print("Panjang vector x= ", len(x_vector))

#tampilkan ukuran vector x
print("Ukuran vector x= ", x_vector.shape)

#tampilkan tipe data dari x
print("Tipe data x=", type(x_vector))

# Akses elemen pertama dari vector x
print("Elemen pertama vector x =", x_vector[0])

#tampilkan tipe data elemen pertama dari vector x
print("Tipe data elemen pertama =", type(x_vector[0][0]))

#lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x_vector.T
print("Transpose vector x= ", x_t)

#tampilkan ukuran dari vector x_t
print("Ukuran vector x_t= ", x_t.shape)

#lakukan transpose pada vector y dan simpan pada variabel y_t
y_vector = np.array([[44],[55],[66],[77]])
print("y=", y_vector)
y_t = y_vector.T
print("Transpose vector y=", y_t)

#tampilkan ukuran dari vector y_t
print("Ukuran vector y_t= ", y_t.shape)
```

```python
import torch

#buat vector x dengan ukuran 3x1 dengan pytorch
xpt_vector = torch.tensor([[22], [44], [66]])
print("x= ", xpt_vector)

#lakukan transpose pada vector x dan simpan pada variabel x_t dengan pytorch
x_t = xpt_vector.t()
print("Transpose vector x= ", x_t)

#tampilkan ukuran dari vector x_t dengan pytorch
print("Ukuran vector x_t= ", x_t.size())

#lakukan transpose pada vector y dan simpan pada variabel y_t dengan pytorch
ypt_vector = torch.tensor([[66], [77], [88], [99]])
print("y=", ypt_vector)
y_t = ypt_vector.t()
print("Transpose vector y=", y_t)

#tampilkan ukuran dari vector y_t dengan pytorch
print("Ukuran vector y_t= ", y_t.size())
```

```python
import tensorflow as tf

#buat vector x dengan ukuran 3x1 dengan tensorflow
xtf_vector = tf.constant([[22], [33], [55]])
print("x= ", xtf_vector)

#lakukan transpose pada vector x dan simpan pada variabel x_t dengan tensorflow
x_t = tf.transpose(xtf_vector)
print("Transpose vector x= ", x_t)

#tampilkan ukuran dari vector x_t dengan tensorflow
print("Ukuran vector x_t= ", x_t.shape)

#lakukan transpose pada vector y dan simpan pada variabel y_t dengan tensorflow
ytf_vector = tf.constant([[1], [0], [0], [1]])
print("y=", ytf_vector)
y_t = tf.transpose(ytf_vector)
print("Transpose vector y=", y_t)

#tampilkan ukuran dari vector y_t dengan tensorflow
print("Ukuran vector y_t= ", y_t.shape)
```

Norm and Unit Vectors

```python
import numpy as np

# buat vector x dengan ukuran 3x1
x_norm = np.array([[11],[22],[33]])
print("Vector x = ", x_norm)
print(x_norm)

# hitung nilai L2 norm dari vector x
norm_x = np.linalg.norm(x_norm)
print("L2 norm dari vector x =", norm_x)

# buat vector v dengan elemen (3,4)
v = np.array([3, 4])
print("Vector v =", v)

# hitung nilai L2 norm dari vector v
norm_v = np.linalg.norm(v)
print("L2 norm dari vector v =", norm_v)

# tampilkan unit vector dari vector v
unit_v = v / norm_v
print("Unit vector dari v =", unit_v)

# hitung panjang unit vector dari vector v
panjang_v = np.linalg.norm(unit_v)
print("Panjang unit vector v =", panjang_v)
```

Basis, Orthogonal, and Orthonormal Vectors

```python
#buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
print("Vector i =", i)
j = np.array([0, 1])
print("Vector j =", j)

#buktikan bahwa vector i dan vector j itu orthogonal
dot_ij = np.dot(i, j)
print("Dot product i · j =", dot_ij)
print("Vector i dan vector j orthogonal karena dot product nya adalah 0")

#Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
print("Vector v =", v)
w = np.array([2, -1])
print("Vector w =", w)
dot_vw = np.dot(v, w)
print("Dot product v · w =", dot_vw)
print("Vector v dan vector w orthogonal karena dot product nya adalah 0")
```

Matrix Tensors

```python
import numpy as np

#buat matrix X berukuran 2x3 menggunakan numpy
X = np.array([[11, 22, 33],[44, 55, 66]])
print("Matrix X =")
print(X)

# tampilkan ukuran dari matrix X
print("Ukuran matrix X =", X.shape)

# tampilkan seluruh elemen dari kolom pertama
print("Kolom pertama matrix X =", X[:,0])

# tampilkan seluruh elemen dari baris kedua
print("Baris kedua matrix X =", X[1, :])
```

```python
import torch

#buat matrix X berukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[22, 0], [11, 22], [55, 55]])
print(X_pt)

# tampilkan ukuran dari matrix X dengan pytorch
print("Ukuran matrix X =", X_pt.shape)

# tampilkan seluruh elemen dari kolom pertama dengan pytorch
print("Kolom pertama matrix X =", X_pt[:,0])

# tampilkan seluruh elemen dari baris kedua dengan pytorch
print("Baris kedua matrix X =", X_pt[1, :])
```

```python
import tensorflow as tf

# buat matrix X_tf berukuran 2x3
X_tf = tf.constant([[22, 44, 66],[11, 33, 55]])

print("Matrix X_tf =")
print(X_tf)

# tampilkan ukuran dari matrix X_tf
print("Ukuran matrix X_tf =", X_tf.shape)

# tampilkan seluruh elemen dari kolom pertama
print("Kolom pertama matrix X_tf =", X_tf[:, 0])


# tampilkan seluruh elemen dari baris kedua
print("Baris kedua matrix X_tf =", X_tf[1, :])
```
